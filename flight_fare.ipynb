```python
# import required modules and library
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

```

```python
# READ DATA FROM EXCEL FILE
train_data = pd.read_excel(r"Data_Train.xlsx")
```

```python
# NOW WE ANALYSIS OUR DATA
```

```python
train_data.head()
```

```python
train_data.info() #SUMMARY OF DATABASE
```

```python
train_data["Duration"].value_counts()#RETURN COUNT OF UNIQE ROW 
```

```python
train_data.dropna(inplace = True)
```

```python
train_data.isnull().sum()
```

```python
#HERE DATE_OF_JOURNEY IS OBJECT DATA TYPE.SO WE CONVERT DATA TYPE INTO TIME STAMP SO WE CAN USE TO TRAIN MODEL
```

```python
# EXTRACT JOURNEY DAY AND MONTH
train_data["Journey_day"] = pd.to_datetime(train_data.Date_of_Journey, format = "%d/%m/%Y").dt.day
train_data["Journey_month"] = pd.to_datetime(train_data["Date_of_Journey"], format = "%d/%m/%Y").dt.month
train_data.head()
```

```python
# WE HAVE CONVERTED DATE_OF_JOURNET COLUMN INTO INTERGERS SO WE DROP IT                     
train_data.drop(["Date_of_Journey"], axis = 1, inplace = True)
```

```python
# SIMILIAR TO DATE_OF_JOURNET WE CAN EXTRACT FROM DEP_TIME
train_data["Dep_Time"] = pd.to_datetime(train_data["Dep_Time"], format="%H:%M")
train_data["Dep_hour"] = pd.to_datetime(train_data["Dep_Time"]).dt.hour
train_data["Dep_min"] = pd.to_datetime(train_data["Dep_Time"]).dt.minute
train_data.drop(["Dep_Time"], axis = 1, inplace = True)
```

```python
train_data.head()
```

```python
# SIMILEARLY FROM ARRIVAL HOUR WE EXTRACT DATE AND TIME
train_data["Arrival_hour"] = pd.to_datetime(train_data["Arrival_Time"]).dt.hour
train_data["Arrival_min"] = pd.to_datetime(train_data["Arrival_Time"]).dt.minute
train_data.drop(["Arrival_Time"], axis=1, inplace=True)
```

```python
train_data.head()
```

```python
# NOW WE FIND DURATION OF PLANE TO TIME TAKEN FROM SOURCE TO DESTINATION
duration = list(train_data["Duration"])
#IN DURATION WE HAVE DATA LIKE ONLY HOURS,ONLY MINUTE AND HOURS AND MINUTE BOTH
#SO WE CONVERT ALL DATA INTO SAME FORMATE HOURS AND MINUTE
for i in range(len(duration)):           
    if len(duration[i].split()) != 2:  
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"  
        else:
            duration[i] = "0h " + duration[i]          

duration_hours = []
duration_mins = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))    # Extract hours from duration
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))   # Extracts only minutes from duration
```

```python
# Adding duration_hours and duration_mins list to train_data dataframe

train_data["Duration_hours"] = duration_hours
train_data["Duration_mins"] = duration_mins
```

```python
train_data.drop(["Duration"], axis = 1, inplace = True)
```

```python
train_data.head()
```

```python
#NOW WE HAVE DIFFRENT AIRLINE SO WE CONVERT CATAGORICAL DATA INTO INDICATER VARIABLE
```

```python
train_data["Airline"].value_counts()
```

```python
Airline = train_data[["Airline"]]
Airline = pd.get_dummies(Airline, drop_first= True)
Airline.head()
```

```python
#SIMILIARLY DOING FO SOURCE
train_data["Source"].value_counts()
```

```python
Source = train_data[["Source"]]
Source = pd.get_dummies(Source, drop_first= True)
Source.head()
```

```python
#SIMILIARLY FOR DESTINATION
train_data["Destination"].value_counts()
```

```python
Destination = train_data[["Destination"]]
Destination = pd.get_dummies(Destination, drop_first = True)
Destination.head()
```

```python
# IN ADDITIONAL_INFO COLUMN WE HAVE MAJORITY DATA OF NO INFO SO WE DROP IT
# ROUTE CAN CONSIDER THREW NO OF STOPS
#SO WE DELETE BOTH COLUMN

train_data.drop(["Route", "Additional_Info"], axis = 1, inplace = True)
```

```python
train_data["Total_Stops"].value_counts()
```

```python
#IN TOTAL_STOP WE HAVE STRING DATA SO WE CONVERT IT TO INTEGER DATA

train_data.replace({"non-stop": 0, "1 stop": 1, "2 stops": 2, "3 stops": 3, "4 stops": 4}, inplace = True)
```

```python
train_data.head()
```

```python
#NOW WE CONCAT ALL THE DUMMY CONVERTED DATA WITH OUR MAIN DATASET

data_train = pd.concat([train_data, Airline, Source, Destination], axis = 1)
```

```python
data_train.head()
```

```python
data_train.drop(["Airline", "Source", "Destination"], axis = 1, inplace = True)
```

```python
data_train.head()
```

```python
data_train.shape
```

```python
# FINALLY WE HAVE DATA REDY TO TRAIN MODEL
```

```python
#DEFINE X
x=data_train.drop(["Price"],axis=1)
```

```python
x.columns
```

```python
x.shape
```

```python
#DEFINE Y
y=data_train["Price"]
```

```python
y.shape
```

```python
y.head()
```

```python
#NOW WE SPLITE DATA INTO TRAINING AND TESTING
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2, random_state=42)
print(x.shape)
print(y.shape)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#HERE WE TRAIN THE MODEL
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
# Y=MX+C
# VALUE OF M
model.coef_
```

```python
# VALUE OF C
model.intercept_
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
